```python
from pytube import YouTube
import pandas as pd
```

```python
import re
from pytube import Playlist
playlist = Playlist('https://www.youtube.com/playlist?list=PLFXhAdtWBlDEpPmJhYw6PyN3vHaUBpnyJ')   
DOWNLOAD_DIR = '/Volumes/NEIRA/AplicacionesTendenciasAD/video_test'
playlist._video_regex = re.compile(r"\"url\":\"(/watch\?v=[\w-]*)")    
print(len(playlist.video_urls))    
for url in playlist.video_urls:
    print(url)    
for video in playlist.videos:
    print('downloading : {} with url : {}'.format(video.title, video.watch_url))
    video.streams.\
        filter(type='video', progressive=True, file_extension='mp4').\
        order_by('resolution').\
        desc().\
        first().\
        download(DOWNLOAD_DIR)
```

```python
import os
ejemplo_dir = '/Volumes/NEIRA/AplicacionesTendenciasAD/video_test'
contenido = os.listdir(ejemplo_dir)
videos = []
for fichero in contenido:
    if os.path.isfile(os.path.join(ejemplo_dir, fichero)) and fichero.endswith('.mp4'):
        videos.append(fichero)
print(videos)
```

```python
lines = videos
with open('/Users/pabloneira/AplicacionesTendenciasAD/name_test.txt', 'w') as f:
    for line in lines:
        f.write(line)
        f.write('\n')
```
